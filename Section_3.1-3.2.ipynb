```python
import pandas as pd
import numpy as np

data = [15,14,13,16,17,14,14,15,12,6,20,14,13,14]
```

```python
df = pd.DataFrame(data)
```

```python
df.describe()
```

You can get the individual items from the DataFrame object.

```python
df.mean()
```

```python
df.var()
```

```python
df.std()
```

We discussed modes in class

```python
df.mode()
```

The coefficient of variance would be given by:

```python
cv = (df.std() / df.mean()) * 100
cv
```

So the coefficient of variance is 21.67%
