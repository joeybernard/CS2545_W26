## Boxplots
- the length of the box represents the IQR
- whiskers show spread and variability
- can show skewness and symmetry

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

minimum = 42000
q1 = 54000
median = 65000
q3 = 92000
maximum = 140000

def format_ticks(value, tick_number):
    return f'{value:,.0f}'

plt.boxplot([[minimum, q1, median, q3, maximum]], vert=False, labels=['Salaries'], widths=0.1)
plt.xlabel('Salaries')
plt.ylabel('Boxplot Showing Distribution of Salaries')
plt.gca().xaxis.set_major_formatter(ticker. FuncFormatter(format_ticks))
```

## Pareto Charts

For categorical data

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Failure Category': ['Cracked Screen', 'Power Button', 'Scratches', 'Does Not Charge', 'Volume Controls'],
    'Frequency': [1348, 739, 1543, 1410, 595]
}

df = pd.DataFrame(data)
df_sorted = df.sort_values(by='Frequency', ascending=False)
df_sorted['Cumulative Percentage'] = (df_sorted['Frequency'].cumsum() / df_sorted['Frequency'].sum()) * 100

fig, ax1 = plt.subplots()

ax1.bar(df_sorted['Failure Category'], df_sorted['Frequency'])
ax1.set_ylabel('Frequency')

ax2 = ax1.twinx()
ax2.plot(df_sorted['Failure Category'], df_sorted['Cumulative Percentage'], color='blue', marker='o')
ax2.set_ylabel('Cumulative Percentage (%)')

ax1.set_xticklabels(df_sorted['Failure Category'], rotation=45, ha='right')

plt.title('Pareto Chart of Smartphone Manufacturing Defects')
plt.show()
```

## Binomial Distribution

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom

n = 20
p = 0.1

x = np.arange(0, n+1)
probabilities = binom.pmf(x, n, p)

plt.bar(x, probabilities)

plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution ($n = 20$, $p = 0.1$)')

plt.show()
```

## Poisson Distributions

```python
import matplotlib.pyplot as plt
from scipy.stats import poisson

mu = 3

x = np.arange(0, 12)
probabilities = poisson.pmf(x, mu)

plt.bar(x, probabilities)

plt.xlabel('Number of Occurrences')
plt.ylabel('Probability')
plt.title('Poisson Distribution ($\mu$ = 3)')

plt.show()
```

## Normal Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu = 120
sigma = 20

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y)
plt.xlabel('Blood Pressure')
plt.ylabel('Probability Density')
plt.title('Normal Distribution for Blood Pressure')

plt.show()
```

## Heatmaps

```python
import seaborn as sns
import matplotlib.pyplot as plt

heatmap_data = sns.load_dataset('flights')
heatmap_data_array = heatmap_data.pivot(index='year', columns='month', values='passengers')

sns.heatmap(heatmap_data_array, cmap='coolwarm')

plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Heatmap of Number of Passengers')

plt.show()
```

## 3D Plotting

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import matplotlib.cm as cm
import matplotlib.colors as mcolors
from matplotlib.patches import Rectangle

flower = sns.load_dataset('iris')

fig = plt.figure(figsize=(9,8), facecolor='white')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('white')

scatter = ax.scatter(flower['sepal_length'], flower['sepal_width'], flower['petal_length'],
                    c=flower['species'].astype('category').cat.codes, cmap='viridis', marker='o')

ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.set_zlabel('Petal Length (cm)')
ax.set_title('3D Scatterplot of Iris Dataset by Species')

species_names = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}

norm = plt.Normalize(flower['species'].astype('category').cat.codes.min(),
                    flower['species'].astype('category').cat.codes.max())
sm = plt.cm.ScalarMappable(cmap='viridis', norm=norm)
sm.set_array([])

cbar = plt.colorbar(sm, ticks=[0, 1, 2], label='Species', fraction=0.02, pad=0.1)
cbar.ax.tick_params(labelsize=10)

fig.subplots_adjust(left=0.1, right=0.85, top=0.9, bottom=0.1)

ax.legend(handles=scatter.legend_elements()[0],
         labels=[species_names[name] for name in flower['species'].astype('category').cat.codes.unique()],
         title='Species')

plt.show()
```
