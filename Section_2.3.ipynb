## Regex

```python
import re
initial_text = "This is a text about Python programming"

pattern = "Python"

words = re.findall(pattern, initial_text)
print(words)
```

```python
pattern2 = "P*"
words2 = re.findall(pattern2, initial_text)
words2
```
