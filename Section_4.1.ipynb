# Statistical Inference and Confidence Intervals

When your data is normally distributed and you know the population standard deviation, you can get the margin of error by scaling the z-score for the percentage of interest by $\frac{\sigma}{\sqrt{n}}$.

In most cases, you do not know this so we have to use the t-distribution. This is the same distribution used in the Student t-test. The "Student" part is the pdeudonym of the author (William Sealy Gosset, from Guinness Brewery) of a paper published in 1908. It was actually developed in 1876 by Helmert and Luroth.

The t-distribution is generated by

$\displaystyle f(t)={\frac {\Gamma {\left({\frac {\nu +1}{2}}\right)}}{{\sqrt {\pi \nu }}\,\Gamma {\left({\frac {\nu }{2}}\right)}}}\left(1+{\frac {t^{2}}{\nu }}\right)^{-(\nu +1)/2}$

This is why we use lookup tables or a computer function. The degrees of freedom is equal to $n-1$, so tied to the number of samples. As the degrees of freedom grow, we get closer to the normal distribution. 

```python
from scipy.stats import t

# What probability do we want for our confidence interval
area_to_left = 0.975
df = 15

# What would the critical value be
round(t.ppf(area_to_left, df), 3)
```

The margin of error would then be

$E = t_c \frac{s}{\sqrt{n}}$

## Confidence Intervals

There are functions available to get confidence intervals for normal distributions, t-distributions and binomial distributions.

For example, you can use the following to get a confidence interval for a normal distribution

```python
import scipy.stats as stats
import numpy as np
import math

sample_mean = 12.5
population_std = 6.3
sample_size = 50

confidence_level = 0.90

standard_error = population_std / math.sqrt(sample_size)

stats.norm.interval(confidence_level, sample_mean, standard_error)
```
